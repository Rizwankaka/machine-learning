```python
# import the libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# import the iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

```python
# label encoding the species in iris dataset
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
iris['species'] = le.fit_transform(iris['species'])
iris.head()
```

```python
# draw a heatmap on iris dataset
sns.heatmap(iris.corr(), annot=True)
```

```python
# draw a pairplot on iris dataset
sns.pairplot(iris)
```

```python
# draw a scatterplot between petal_length and petal_width
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=iris)
```

```python
# draw a density plot between sepal_length and petal_length
sns.kdeplot(x='sepal_length', y='petal_length', hue='species', data=iris)
```

```python
# draw a boxenplot between sepal_width and petal_width
sns.boxenplot(x='sepal_length', y='petal_length', hue='species', data=iris)
```

```python
# find the null values in the iris dataset
iris.isnull().sum()
```

```python
# find the outliers in the iris dataset
iris.describe()
```

```python
# apply the standard scalar in the dataset 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
iris_std = sc.fit_transform(iris)
iris_std
```
