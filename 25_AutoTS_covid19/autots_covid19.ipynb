# Covid-19 deaths prediction model 

```python
import pandas as pd 
import numpy as np 
```

```python
# load the dataset
df = pd.read_csv('COVID19_India.csv') # https://raw.githubusercontent.com/amankharwal/Website-data/master/COVID19%20data%20for%20overall%20INDIA.csv
df.head()
```

```python
# find null values 
df.isnull().sum().sort_values(ascending=False)
```

```python
# drop data column
df.drop(['Date'], axis=1, inplace=True)
```

```python
df.columns
```

```python
# plot in plotly 
import plotly.express as px 
fig= px.bar(df, 'Date_YMD', y='Daily Confirmed')
fig.show()
```

```python
cases= df['Daily Confirmed'].sum()
deceased=df['Daily Deceased'].sum()

labels=['Confirmed', 'Deceased']
values=[cases, deceased]

fig=px.pie(df, values=values, 
           names=labels,
           title='Daily Confirmed Cases vs Daily Deaths', hole=0.5)
fig.show()
```

```python
death_rate=deceased/cases*100
death_rate
```

```python
import plotly.express as px
fig=px.bar(df, x='Date_YMD', y='Daily Deceased')
fig.show()
```

## AUTOTS library

```python
from autots import AutoTS
```

```python
# create a model 
model = AutoTS(forecast_length=5, frequency='infer', ensemble='simple')

# fit the model 
model= model.fit(df, date_col='Date_YMD', value_col='Daily Deceased', id_col=None)

# predict the future values
prediction = model.predict()
forecast= prediction.forecast
print(forecast)
```
