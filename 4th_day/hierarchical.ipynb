## Hierarchical Clustering
hierarchical clustering is a method for grouping data into disjoint subsets of similar data. It is a non-parametric method that is used 
when the number of clusters is not known in advance. It is also used to find clusters in data that is not normally distributed.

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch 
from sklearn.datasets import make_blobs

# Generate a random dataset with 100 samples and 3 features 
X, y = make_blobs(n_samples=100, n_features=3, centers=3, random_state=0)

# Perform hierarchical clustering with the AgglomerativeClustering algorithm 
clustering = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
clustering.fit(X)

# Plot the dendrogram 
plt.figure(figsize=(10,7))
plt.title('Dendogram')
dendrogram=sch.dendrogram(sch.linkage(X, method='ward'))
plt.show()
```

## Make dendrogram on tips dataset

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
from sklearn.datasets import make_blobs

df= sns.load_dataset('tips')
X= df[['total_bill', 'size', 'tip']]

# Perform hierarchical clustering with the AgglomerativeClustering algorithm 
clustering = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
clustering.fit(X)

# Plot the dendrogram 
plt.figure(figsize=(10,7))
plt.title('Dendogram')
dendrogram=sch.dendrogram(sch.linkage(X, method='ward'))
plt.show()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage 

# Load the tips dataset from seaborn 
tips = sns.load_dataset('tips')

# select relevent features for clustering 
features=['total_bill', 'tip']

# Extract the selected features from the dataset
X = tips[features]

# Perform Agglomerative Clustering
n_cluster=5 # Number of clusters
agg_clustering=AgglomerativeClustering(n_clusters=n_cluster)
agg_labels=agg_clustering.fit_predict(X)

# Add cluster labels to the dataset
tips['cluster']=agg_labels

# Visualize the clusters 
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='cluster', palette='viridis')
plt.title('AgglomerativeClustering on tips dataset')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()

# Plot the dendrogram 
dendrogram(linkage(X, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd 
from scipy.cluster.hierarchy import dendrogram, linkage 

# Load the tips dataset from seaborn 
tips = sns.load_dataset('tips')

# Select the relevent features for clustering 
features=['total_bill', 'tip', 'size']

# Extract the selected features from the dataset
X = tips[features]

# Compute the linkage matrix
linkage_matrix=linkage(X, method='ward')

# Plot the dendrogram
plt.figure(figsize=(10,7))
dendrogram(linkage_matrix)  
plt.title('Dendogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()

# Plot the heatmap
plt.figure(figsize=(8,6))
sns.heatmap(X, cmap='coolwarm')
plt.title('Heatmap')
plt.xlabel('features')
plt.ylabel('Samples')
plt.show()
```
