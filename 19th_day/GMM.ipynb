# **Gaussian Mixture Model - GMM**

```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.mixture import GaussianMixture
```

```python
# Load the iris dataset
iris = datasets.load_iris()
X = iris.data[:, :2] # We will use only the first two features for simplicity 

# Create a Guassian Mixture Model with 3 components/clusters
gmm = GaussianMixture(n_components=3)

# Fit the model
gmm.fit(X)

# Predict the cluster label for each data point
labels = gmm.predict(X)

# Visualize the clustering results 
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('Guassian Mixture Model Clustering on Iris Dataset')
plt.show()
```

```python
# Load the iris dataset
iris = datasets.load_iris()
X = iris.data[:, :2] # We will use only the first two features for simplicity 

# Create a Guassian Mixture Model with 3 components/clusters
gmm = GaussianMixture(n_components=3)

# Fit the model
gmm.fit(X)

# Predict the cluster label for each data point
labels = gmm.predict(X)

# Compute the log-likelihood of the data # evaluation metric, shows how well the clusters are formed
log_likelihood = gmm.score(X)

# Print the log-likelihood
print('Log-likelihood',log_likelihood)

# Visualize the clustering results 
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('Guassian Mixture Model Clustering on Iris Dataset')
plt.show()
```

```python
# Load the iris dataset
iris = datasets.load_iris()
X = iris.data[:, :2] # We will use only the first two features for simplicity 

# Create a Guassian Mixture Model with 3 components/clusters
gmm = GaussianMixture(n_components=3)

# Fit the model
gmm.fit(X)

# Predict the cluster label for each data point
labels = gmm.predict(X)

# Compute the log-likelihood of the data # evaluation metric, shows how well the clusters are formed
log_likelihood = gmm.score(X)

# Compute the AIC and BIC
n_samples,_=X.shape
aic = gmm.aic(X) # AIC (Akaike Information Criterion)
bic = gmm.bic(X) # BIC (Bayesian Information Criterion)

# Print the evaluation metrics
print('Log-likelihood',log_likelihood)
print('AIC',aic)
print('BIC',bic)

# Visualize the clustering results 
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('Guassian Mixture Model Clustering on Iris Dataset')
plt.show()
```

- near to zero AIC value is considered good. Otherwise, it shows over or underfit the model
- similar to AIC, BIC should also be less. high value means cluster formed here in which probability didn't fit properly.
